```python
df_predictions = pd.read_csv(utils.latest_smart_prediction(), index_col=0)
```

```python
df_predictions.head()
```

```python
outcome = df_predictions['true_change']
```

```python
df_predictions = df_predictions.drop(columns="true_change")
```

```python
df_predictions = df_predictions.drop(columns="pow10")
```

```python
df_predictions = df_predictions.drop(columns="two_hops")
```

```python
df_predictions.head()
```

```python
values = []
for col in df_predictions.columns:
    column = df_predictions[col]
    correct = column.eq(outcome).sum()
    incorrect = column.ne(outcome).sum() - column.eq(-1).sum()
    tpr = round(correct / len(column), 6)
    fpr = round(incorrect / len(column), 6)
    print(tpr, "&", fpr, col)
    values.append((fpr, tpr))
```
